<h2>Plotly - Exercício 001</h2>

```python
# Importa o módulo plotly.express, usado para criar gráficos de forma simples e rápida
import plotly.express as px

# Importa plotly.graph_objects, usado para gráficos mais customizados (não é usado diretamente aqui)
import plotly.graph_objects as go

# Carrega o dataset "gapminder" que já vem pronto dentro do plotly.express
# Esse dataset contém dados de países (PIB, expectativa de vida, população, etc.)
df = px.data.gapminder()
#df.to_csv("gapminder.csv", index=False)
df.to_excel("gapminder.xlsx", index=False) #exporta dataset imbutido na biblioteca

# Cria um gráfico de dispersão (scatter plot)
fig = px.scatter(
    df,                    # DataFrame que contém os dados
    x='gdpPercap',          # Coluna usada no eixo X (PIB per capita)
    y='lifeExp',            # Coluna usada no eixo Y (expectativa de vida)
    size='pop',             # Tamanho dos pontos proporcional à população
    color='continent',      # Cor dos pontos definida pelo continente
    hover_name='country',   # Nome do país aparece ao passar o mouse
    log_x=True,             # Usa escala logarítmica no eixo X
    template='plotly_white',# Define um tema visual com fundo branco
    title='Gapminder 2007: PIB per capita vs Expectativa de Vida'
                             # Título do gráfico
)

# Importa o módulo plotly.io, responsável pelos métodos de renderização e exportação
import plotly.io as pio

# Define que o gráfico será aberto no navegador padrão do sistema
pio.renderers.default = "browser"

# Exporta o gráfico para um arquivo HTML
# include_plotlyjs='cdn' carrega a biblioteca Plotly via CDN (arquivo menor)
# full_html=True cria um HTML completo e independente
fig.write_html("gapminder7.html", include_plotlyjs='cdn', full_html=True)

# Exibe o gráfico (abre no navegador, conforme o renderer configurado)
fig.show()
```

```python
import pandas as pd
from plotly.subplots import make_subplots
import plotly.graph_objects as go

Capitais = ['Porto Alegre', 'Florianópolis', 'Curitiba']
Populacao = [1388794, 587486, 1830795]
populacao_total = sum(Populacao)
populacao_percent = [
    Populacao[0]/populacao_total,
    Populacao[1]/populacao_total,
    Populacao[2]/populacao_total
]

tabela = {
    'Capitais': Capitais,
    'Percentual_Populacional': populacao_percent
}

df_tabela = pd.DataFrame(tabela)
# Painel com 2 colunas
painel = make_subplots(
    rows=1,
    cols=2,
    specs=[[{"type": "table"}, {"type": "domain"}]]
)

# 📊 TABELA (CORRETO)
painel.add_trace(
    go.Table(
        header=dict(values=list(df_tabela.columns)),
        cells=dict(values=[df_tabela[col] for col in df_tabela.columns])
    ),
    row=1,
    col=1
)

# 🥧 PIZZA
painel.add_trace(
    go.Pie(
        labels=df_tabela['Capitais'],
        values=df_tabela['Percentual_Populacional'],
        name="População %"
    ),
    row=1,
    col=2
)
```
